```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
retail_df=pd.read_csv('Warehouse_and_Retail_Sales.csv')
retail_df.head()
```
